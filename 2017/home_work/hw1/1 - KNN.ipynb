```python
from matplotlib.colors import ListedColormap
from sklearn import model_selection, datasets, metrics, neighbors 

import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = datasets.make_classification(n_samples=100, n_features =2, n_informative = 2, 
                                                      n_classes = 3, n_redundant=0, 
                                                      n_clusters_per_class=1, random_state=3)
```

```python
colors = ListedColormap(['red', 'blue', 'yellow'])
light_colors = ListedColormap(['lightcoral', 'lightblue', 'lightyellow'])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], 
            c=data[1], cmap=colors, s=100)
```

```python
scores = []
for k in np.arange(1, 10):
    kfold = model_selection.KFold(n_splits=5)
    clf = neighbors.KNeighborsClassifier(n_neighbors=k)
    scores.append(model_selection.cross_val_score(clf, data[0], data[1], scoring="accuracy", cv=kfold).mean())
    
plt.plot(np.arange(1, 10), scores)
```
