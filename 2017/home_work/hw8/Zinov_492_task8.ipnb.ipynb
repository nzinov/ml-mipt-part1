```python
import numpy as np
import pylab as plt
%matplotlib inline
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    #Sigma = 1.0 / l * F * np.transpose(F)
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(F)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot for i in range(n)])
    m = np.searchsorted(var_i, varRetained) + 1
    print('{:.2} variance retained in {} dimensions'.format(var_i[m - 1], m))

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:m]
    G = V

    return G, U_reduced
```

```python
import skimage
import skimage.io
```

```python
im = skimage.io.imread("two-lorikeet-birds-2293918.jpg", dtype="float64")
```

```python
plt.imshow(im)
```

```python
colors = skimage.color.rgb2lab(im).reshape((-1, 3)) #В пространстве lab хорошо работает евклидова метрика
```

```python
from sklearn.cluster import KMeans, MeanShift, estimate_bandwidth, AgglomerativeClustering
import skimage.measure
bandwidth = estimate_bandwidth(X, quantile=0.005, n_samples=500)
clust = MeanShift(bandwidth=bandwidth, bin_seeding=True)
```

```python
labels = clust.fit_predict(colors)
```

```python
new_colors = clust.cluster_centers_[labels]
```

```python
new_image = skimage.color.lab2rgb(new_colors.reshape(im.shape))
```

```python
print("MeanShift, SSIM: {}".format(skimage.measure.structural_similarity(im, new_image, multichannel=True)))
plt.imshow(new_image)
print(set(labels))
```

```python
for c_num in [2, 3, 10, 20]:
    print("Число кластеров: {}".format(c_num))
    clust = KMeans(c_num)
    labels = clust.fit_predict(colors)
    new_colors = clust.cluster_centers_[labels]
    new_image = skimage.color.lab2rgb(new_colors.reshape(im.shape))
    print("SSIM: {}".format(skimage.measure.structural_similarity(im, new_image, multichannel=True)))
    plt.imshow(new_image)
    plt.show()
```

Чем больше цветов осталять - тем лучше (логично). При 10 цветах картинка красивая и часть шума (watermark) даже удаляется. MeanShift с 6 цветам немного оьошел KMeans c 10 - классно.

```python
from sklearn.neighbors import kneighbors_graph
knn_graph = kneighbors_graph(colors, 30, include_self=False)

for c_num in [2, 3, 10, 20]:
    print("Число кластеров: {}".format(c_num))
    clust = AgglomerativeClustering(c_num, connectivity=knn_graph)
    labels = clust.fit_predict(colors)
    pallete = np.array([colors[labels == c].mean(axis=0)] for c in labels)
    new_colors = np.array([pallete[c] for c in labels])
    new_image = skimage.color.lab2rgb(new_colors.reshape(im.shape))
    print("SSIM: {}".format(skimage.measure.structural_similarity(im, new_image, multichannel=True)))
    plt.imshow(new_image)
    plt.show()
```
